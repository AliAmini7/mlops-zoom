```python
import pandas as pd
```

```python
df_jan = pd.read_parquet('yellow_tripdata_2022-01.parquet')
```

```python
# number of the columns in the df_jan dataframe
len(df_jan.columns)
```

```python
# duration of the trips in minutes
df_jan['duration'] = df_jan['tpep_dropoff_datetime'] - df_jan['tpep_pickup_datetime']
```

```python
# change df_jan['duration'] to minutes
df_jan['duration'] = df_jan['duration'].apply(lambda x: x.total_seconds()/60)
```

```python
# standard deviation of the df_jan.duration
df_jan.duration.std()
```

```python
# fraction left after dropping the outliers
len(df_jan[(df_jan.duration >=1) & (df_jan.duration <= 60)])/len(df_jan)*100
```

```python
df_jan = df_jan[(df_jan.duration >=1) &  (df_jan.duration <= 60)]
```

```python
df_jan[categorical].dtypes
```

```python
categorical = ['PULocationID', 'DOLocationID']

X_train_df = df_jan[categorical].astype(str)
y_train = df_jan['duration'].values
```

```python
y_train, X_train_df
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer()
```

```python
train_dict = X_train_df.to_dict(orient = 'records')
```

```python
train_dict
```

```python
X_train = dv.fit_transform(train_dict)
```

```python
# number of the columns of the sparse X_train matrix
X_train.shape[1]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# RMSE for the 
mean_squared_error(y_train, y_pred, squared=False)
```

```python
df_feb = pd.read_parquet('yellow_tripdata_2022-02.parquet')
```

```python
df_feb.head()
```

```python
# duration of the trips in minutes
df_feb['duration'] = df_feb['tpep_dropoff_datetime'] - df_feb['tpep_pickup_datetime']
df_feb['duration'] = df_feb['duration'].apply(lambda x: x.total_seconds()/60)
```

```python
df_feb = df_feb[(df_feb.duration >= 1) &  (df_feb.duration <= 60)]
```

```python
categorical = ['PULocationID', 'DOLocationID']

X_val_df = df_feb[categorical].astype(str)
y_val = df_feb['duration'].values
val_dict = X_val_df.to_dict(orient = 'records')

X_val = dv.transform(val_dict)

```

```python

y_pred = lr.predict(X_val)

# RMSE for the 
mean_squared_error(y_val, y_pred, squared=False)
```
